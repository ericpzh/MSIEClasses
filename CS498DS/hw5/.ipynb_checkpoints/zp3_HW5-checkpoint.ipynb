```python
import csv
import pandas as pd
import numpy as np

from sklearn.model_selection import KFold
#from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv('HTRU_2.csv')
df.head()
```

```python
def svm(X_train, y_train,X_test, y_test):
    print("SVM:")
    
    print("0.1:")
    clf = LinearSVC(random_state=0, tol=1e-5, C=0.1)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))

    result = [[accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))]]

    print("1:")
    clf = LinearSVC(random_state=0, tol=1e-5, C=1)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))

    result.append([accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))])


    print("10:")
    clf = LinearSVC(random_state=0, tol=1e-5, C=10)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    
    result.append([accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))])


    return result
```

```python
def dt(X_train, y_train,X_test, y_test):
    print("DT:")
    
    print("3:")
    clf = tree.DecisionTreeClassifier(max_depth=3)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    
    result = [[accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))]]

    print("4:")
    clf = tree.DecisionTreeClassifier(max_depth=4)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))

    result.append([accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))])

    print("6:")
    clf = tree.DecisionTreeClassifier(max_depth=6)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    
    result.append([accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))])


    return result
```

```python
def rf(X_train, y_train,X_test, y_test):
    print("RF:")
    print("5:")
    clf = RandomForestClassifier(max_depth=5, n_estimators=5, random_state=0)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    
    result = [[accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))]]
    
    print("11:")
    clf = RandomForestClassifier(max_depth=5, n_estimators=11, random_state=0)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    
    result.append([accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))])

    print("13:")
    clf = RandomForestClassifier(max_depth=5, n_estimators=13, random_state=0)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    
    result.append([accuracy_score(y_test, clf.predict(X_test)), recall_score(y_test, clf.predict(X_test)), precision_score(y_test, clf.predict(X_test)), f1_score(y_test, clf.predict(X_test))])

    return result
```

```python
X, y = df[['MeanIP','StdIP','KurtosisIP','SkewIP','MeanDM','StdDM','KurtosisDM','SkewDM']].to_numpy(), df['Class'].to_numpy()

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

kf = KFold(n_splits=5)
kf.get_n_splits(X)

svmr = [[[],[],[],[]],[[],[],[],[]],[[],[],[],[]]] #['acc, pre, re, f1, acc, pre, re, f1, ...']
dtr = [[[],[],[],[]],[[],[],[],[]],[[],[],[],[]]]
rfr = [[[],[],[],[]],[[],[],[],[]],[[],[],[],[]]]
for train_index, test_index in kf.split(X):
    X_train, X_val = X[train_index], X[test_index]
    y_train, y_val = y[train_index], y[test_index]
    
    r = svm(X_train, y_train,X_val, y_val)
    
    svmr[0][0].append(r[0][0])
    svmr[0][1].append(r[0][1])
    svmr[0][2].append(r[0][2])
    svmr[0][3].append(r[0][3])
    svmr[1][0].append(r[1][0])
    svmr[1][1].append(r[1][1])
    svmr[1][2].append(r[1][2])
    svmr[1][3].append(r[1][3])
    svmr[2][0].append(r[2][0])
    svmr[2][1].append(r[2][1])
    svmr[2][2].append(r[2][2])
    svmr[2][3].append(r[2][3])
    
    r = dt(X_train, y_train,X_val, y_val)
    dtr[0][0].append(r[0][0])
    dtr[0][1].append(r[0][1])
    dtr[0][2].append(r[0][2])
    dtr[0][3].append(r[0][3])
    dtr[1][0].append(r[1][0])
    dtr[1][1].append(r[1][1])
    dtr[1][2].append(r[1][2])
    dtr[1][3].append(r[1][3])
    dtr[2][0].append(r[2][0])
    dtr[2][1].append(r[2][1])
    dtr[2][2].append(r[2][2])
    dtr[2][3].append(r[2][3])
    
    r = rf(X_train, y_train,X_val, y_val)
    rfr[0][0].append(r[0][0])
    rfr[0][1].append(r[0][1])
    rfr[0][2].append(r[0][2])
    rfr[0][3].append(r[0][3])
    rfr[1][0].append(r[1][0])
    rfr[1][1].append(r[1][1])
    rfr[1][2].append(r[1][2])
    rfr[1][3].append(r[1][3])
    rfr[2][0].append(r[2][0])
    rfr[2][1].append(r[2][1])
    rfr[2][2].append(r[2][2])
    rfr[2][3].append(r[2][3])
```

```python
print("SVM")
for i in svmr:
    print("new model")
    for j in i:
        print("mean:")
        print(np.mean(j))
        print("std:")
        print(np.std(j))
        
print("DT")
for i in dtr:
    print("new model")
    for j in i:
        print("mean:")
        print(np.mean(j))
        print("std:")
        print(np.std(j))
        
print("RF")
for i in rfr:
    print("new model")
    for j in i:
        print("mean:")
        print(np.mean(j))
        print("std:")
        print(np.std(j))
```
