```python
import csv
import pandas as pd
import numpy as np

#from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('HTRU_2.csv')
df.head()
```

```python
def svm(X_train, y_train,X_test, y_test):
    print("SVM:")
    
    print("0.1:")
    clf = LinearSVC(random_state=0, tol=1e-5, C=0.1)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))

    print("1:")
    clf = LinearSVC(random_state=0, tol=1e-5, C=1)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))

    print("10:")
    clf = LinearSVC(random_state=0, tol=1e-5, C=10)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
```

```python
def dt(X_train, y_train,X_test, y_test):
    print("DT:")
    
    print("3:")
    clf = tree.DecisionTreeClassifier(max_depth=3)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))

    print("4:")
    clf = tree.DecisionTreeClassifier(max_depth=4)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))

    print("6:")
    clf = tree.DecisionTreeClassifier(max_depth=6)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
```

```python
def rf(X_train, y_train,X_test, y_test):
    print("RF:")
    print("5:")
    clf = RandomForestClassifier(max_depth=5, n_estimators=5, random_state=0)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    print("11:")
    clf = RandomForestClassifier(max_depth=5, n_estimators=11, random_state=0)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
    print("13:")
    clf = RandomForestClassifier(max_depth=5, n_estimators=13, random_state=0)
    clf.fit(X_train, y_train)
    print("accuracy:")
    print(accuracy_score(y_test, clf.predict(X_test)))
    print("recall:")
    print(recall_score(y_test, clf.predict(X_test)))
    print("precision:")
    print(precision_score(y_test, clf.predict(X_test)))
```
