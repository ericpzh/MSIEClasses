```python
from copy import deepcopy
from itertools import chain
from collections import Counter

def indicesList(k, l, pattern, tran):
    # indices = []
    # for i in range(l - k):
    #     if pattern == tran[i : i + k]:
    #         indices.append(i + k)
    
    indices = [(i + k) for i in range(l - k) if pattern == tran[i : i + k]]
    return indices

def uniqueItem(listOfLists):
    for item in set(chain(*listOfLists)): yield item

def countElem(listOfLists, elem): 
    return sum(sublist.count(elem) for sublist in listOfLists)


min_sup = 2
trans = docs
# trans = []
# for line in content:
#     trans.append(line.strip().split())
all_elements = list(uniqueItem(trans))
# all_elements.sort()
# pattern = {}
# pattern_list = [[elem] for elem in all_elements if countElem(trans, elem) >= min_sup]
pattern_list = []
for elem in all_elements:
    count = countElem(trans, elem)
    if count >= min_sup:
        pattern_list.append([elem])
# pattern_list = [[key] for key in pattern]
# print(pattern_list1 == pattern_list)

def sequentialContiguousPatternMining(begin, end, trans, one_freq_itemset):
    pattern_dic = {}
    pattern_list = deepcopy(one_freq_itemset)
    reject_sup = 0
    for k in range(begin, end):
        temp_pattern_list = []
        for pattern in pattern_list:
            temp_list = []
            for tran in trans:
                l = len(tran)
                indices = indicesList(k, l, pattern, tran)
                # print(indices)
                if indices:
                    for i in indices:
                        temp_list.append(tran[i])
    
            count_list = Counter(temp_list)
            for elem in set(temp_list):
                count = count_list[elem]
                if count >= min_sup and count >= reject_sup:
                    temp_pattern = deepcopy(pattern)
                    temp_pattern.append(elem)
                    temp_pattern_list.append(temp_pattern)
                    pattern_dic[' '.join(temp_pattern)] = count_list[elem]
                    if len(pattern_dic) > 20:
                        reject_elem = [v[0] for v in sorted(pattern_dic.items(),key = lambda kv: (-kv[1], kv[0]))][-1]
                        reject_sup = pattern_dic[reject_elem]
                        del pattern_dic[reject_elem]
            pattern_list = temp_pattern_list
    ordered_patterns = [v[0] for v in sorted(pattern_dic.items(),key = lambda kv: (-kv[1], kv[0]))]
    return pattern_dic, ordered_patterns

pattern_25, ordered_patterns = sequentialContiguousPatternMining(1, 5, trans, pattern_list)

for elem in ordered_patterns:
    print('[%d, \'%s\']' % (pattern_25[elem], elem))
```
